```python
import tiktoken
from datasets import load_dataset
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
dataset = load_dataset("Arsive/toxicity_classification_jigsaw")
```

```python
df = dataset['train'].to_pandas()
df.head()
```

```python
comment_list = list(df["comment_text"])
len(comment_list)
```

```python
comment_list[:3]
```

```python
comment_strings = " ".join(comment_list)
comment_sents = sent_tokenize(comment_strings)
len(comment_sents)
```

```python
comment_sents[:4]
```

```python
[word_tokenize(s) for s in comment_sents]
```

```python
comment_tokens = word_tokenize(comment_strings)
len(comment_tokens)
```

```python
enc = tiktoken.encoding_for_model("gpt-4")
```

```python
comment_tokens = enc.encode(comment_strings)
len(comment_tokens)
```
